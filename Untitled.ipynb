```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# 
plt.figure(num=None, figsize=(5, 5), dpi=300)

#  Cora
data = [
    [0.9124, 0.9197, 0.9239, 0.9312, 0.9396, 0.9466, 0.94196, 0.93772, 0.93337],  # cora
    [0.91191, 0.9203, 0.9257, 0.9275, 0.9293, 0.9373, 0.93151, 0.9271, 0.92255],  # YST
]

#  x 
x = np.array(['1:1', '1:2', '1:3', '1:4', '1:5', '1:6', '1:7', '1:8', '1:9'])

# 
colors = [(230 / 255, 111 / 255, 81 / 255), (21 / 255, 151 / 255, 165 / 255)]
markers = ['p', '*']
labels = ['Cora', 'YST']

# 
for i in range(len(data)):
    x_new = np.linspace(0, len(x)-1, 300)
    spl = make_interp_spline(np.arange(len(x)), data[i], k=3)
    y_smooth = spl(x_new)
    plt.plot(x_new, y_smooth, color=colors[i], marker=markers[i], label=labels[i], linewidth=1.5, markersize=7)

#  x 
x_ticks = np.array(['1:1', '1:2', '1:3', '1:4', '1:5', '1:6', '1:7', '1:8', '1:9'])
labels = ['0.10', '0.20', '0.30', '0.40', '0.50', '0.60', '0.70', '0.80', '0.90']

#  x 
plt.xlim(-0.25, len(x_ticks)-0.75)
# plt.ylim(0.735, 0.795)

#  y 
start = np.min(data)
end = np.max(data)
plt.xticks(np.arange(len(x_ticks)), labels, fontproperties='Times New Roman', fontsize=15)
y_ticks = np.linspace(0.90, 0.95, 5)

plt.yticks(y_ticks, fontproperties='Times New Roman', fontsize=15)
y_formatter = plt.FormatStrFormatter('%.2f')  # 
plt.gca().yaxis.set_major_formatter(y_formatter)
plt.gca().xaxis.set_tick_params(width=1.25)  # Adjust the x-axis line thickness (width)
plt.gca().yaxis.set_tick_params(width=1.25)
for spine in plt.gca().spines.values():
    spine.set_linewidth(1.25)

plt.xlabel('', fontdict={'family': 'Times New Roman', 'size': 19}, labelpad=1)
plt.ylabel('AUC', fontdict={'family': 'Times New Roman', 'size': 19}, labelpad=1)

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc='upper left', ncol=1,
           frameon=True,
           prop={'family': 'Times New Roman', 'size': 9},
           markerscale=1, bbox_to_anchor=(0.01, 1))

plt.savefig(r'D:\GuoPingXia\Paper\result\hyperparameters\_Cora.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
